接下来，我将提供一个脚本（`scripts/04_generate_qa_instructions.py`），它将读取您处理过的语料库数据（`abstracts_corpus.jsonl` 和 `fulltext_index.jsonl`），并使用 DeepSeek 模型生成“理论问题 + 答案”对。该脚本将结果保存到 `data/processed/qa_instructions.jsonl`。

````python
import jsonlines
import json
import os
from tqdm import tqdm
import re # 导入正则表达式模块
import shutil # 导入 shutil 模块用于文件操作

def generate_qa_pair(paper_content, client): # 将 'model' 改为 'client'
    """使用 DeepSeek API 根据提供的论文内容生成理论问题和答案。"""
    prompt = f"""
基于以下学术论文内容，生成一个理论问题及其对应的详细答案。问题应侧重于核心概念、方法论或发现。输出应为包含 'instruction'、'input'（空字符串）和 'output' 键的 JSON 对象。

论文内容：
```
{paper_content}
```

示例输出格式：
{{"instruction": "ABC 模型的核心假设是什么？", "input": "", "output": "ABC 模型的核心假设是..."}}

现在生成问答对：
"""
    try:
        response = client.chat.completions.create(
            model="deepseek-chat", # 指定 DeepSeek 模型
            messages=[
                {"role": "user", "content": prompt}
            ],
            temperature=0.7, # 可选：调整以平衡创造性和忠实性
            response_format={"type": "json_object"} # 请求 JSON 格式输出
        )
        # 从 DeepSeek API 响应中提取内容
        response_text = response.choices[0].message.content.strip()
        # 如果使用了 response_format 并强制执行，则无需检查 ```json 包装
        qa_data = json.loads(response_text)
        return qa_data
    except Exception as e:
        print(f"生成内容问答时出错: {paper_content[:100]}... 错误: {e}")
        return None


output_dir = 'data/processed'
os.makedirs(output_dir, exist_ok=True)
output_filepath = os.path.join(output_dir, 'qa_instructions.jsonl')

# 定义 Google Drive 上的保存路径 (假设已在之前的单元格中挂载并设置)
drive_output_dir = '/content/drive/MyDrive/or_corpus_data/processed' # 从之前设置的变量获取
os.makedirs(drive_output_dir, exist_ok=True) # 确保 Google Drive 目录存在
drive_output_filepath = os.path.join(drive_output_dir, 'qa_instructions.jsonl')

# 定义 GitHub 仓库的路径和文件路径
github_repo_path = '/content/Datasets-for-Top-Journals-in-Operations-Research-and-Optimization'
github_qa_dir = os.path.join(github_repo_path, 'data/processed')
os.makedirs(github_qa_dir, exist_ok=True) # 确保 GitHub 仓库的 data/processed 目录存在
github_qa_filepath = os.path.join(github_qa_dir, 'qa_instructions.jsonl')


# 加载摘要语料库，使用新的路径
abstracts_filepath = '/content/Datasets-for-Top-Journals-in-Operations-Research-and-Optimization/data/processed/abstracts_corpus.jsonl'
all_papers = []
with jsonlines.open(abstracts_filepath) as reader:
    for obj in reader:
        all_papers.append(obj)

# 加载全文索引，构建 DOI 到全文路径的映射，使用新的路径
fulltext_index_filepath = '/content/Datasets-for-Top-Journals-in-Operations-Research-and-Optimization/data/processed/fulltext_index.jsonl'
fulltext_map = {}
if os.path.exists(fulltext_index_filepath):
    with jsonlines.open(fulltext_index_filepath) as reader:
        for entry in reader:
            filename = entry.get('filename', '')
            path = entry.get('path', '')

            # 从 filename 中提取 DOI
            # 假设 filename 格式为 'Journal_Name_10.xxxx_yyyy.txt'
            # 需要将 filename 中的 '_' 替换为 '/' 来匹配原始 DOI 格式，并提取 DOI 部分
            # 例如: 'Operations_Research_10.1287_opre.2022.2414.txt' -> '10.1287/opre.2022.2414'
            doi_match = re.search(r'(10\.\d{4,}\/[\w\d\.\-]+)\.txt$', filename.replace('_', '/'))
            if doi_match:
                # 构建完整的 DOI 链接，以匹配 all_papers 中的 doi 格式
                original_doi = 'https://doi.org/' + doi_match.group(1)
                if original_doi and path:
                    fulltext_map[original_doi] = path

# 由于 API 成本/速率限制，您可能希望抽取一个更小的子集进行初始测试
# 例如，处理前 50 篇论文：
# papers_to_process = all_papers[:50]
# 或者处理所有论文：
papers_to_process = all_papers

print(f"正在处理 {len(papers_to_process)} 篇论文以生成问答指令...")

qa_instructions = []
# 保存原始工作目录，以便在 git 操作后返回
original_cwd = os.getcwd()

for i, paper in enumerate(tqdm(papers_to_process, desc="生成问答对")):
    content_to_use = paper.get('abstract', '') # 默认使用摘要

    paper_doi = paper.get('doi')
    if paper_doi and paper_doi in fulltext_map: # 如果有 DOI 并且在全文映射中找到
        fulltext_path = fulltext_map[paper_doi]
        # 检查全文文件是否存在并读取内容
        if os.path.exists(fulltext_path):
            try:
                with open(fulltext_path, 'r', encoding='utf-8') as f:
                    fulltext_content = f.read()
                if fulltext_content.strip(): # 如果全文内容不为空
                    content_to_use = fulltext_content # 优先使用全文内容
            except Exception as read_err:
                print(f"读取全文文件 {fulltext_path} 时出错: {read_err}")

    # 限制内容长度以避免 API 输入限制
    # DeepSeek API 的上下文窗口通常较大 (例如 32K 或 128K token)，
    # 但为了稳妥起见并控制成本，我们可以设置一个合理的字符限制。
    # 假设 1个中文字符大约对应 1.5个英文token，这里设置 16000 个字符作为一个经验值。
    MAX_CONTENT_LENGTH = 16000
    if len(content_to_use) > MAX_CONTENT_LENGTH:
        content_to_use = content_to_use[:MAX_CONTENT_LENGTH] + "... [内容已截断]"

    if not content_to_use.strip(): # 如果依然没有有效内容（无论是全文还是摘要）
        # print(f"跳过论文（DOI: {paper_doi}），因为它没有有效内容。")
        continue # 跳过此论文

    qa_pair = generate_qa_pair(content_to_use, deepseek_client) # 使用 DeepSeek 客户端生成问答对
    if qa_pair:
        qa_instructions.append(qa_pair)

    # 每隔 50 条问答对保存一次到本地、Google Drive 和 GitHub
    if (len(qa_instructions) % 50 == 0) and (len(qa_instructions) > 0):
        # 1. 保存到本地
        with jsonlines.open(output_filepath, mode='w') as writer:
            writer.write_all(qa_instructions)
        print(f"\n[Checkpoint] 已保存 {len(qa_instructions)} 条问答指令到本地: {output_filepath}")

        # 2. 复制到 Google Drive
        try:
            shutil.copyfile(output_filepath, drive_output_filepath)
            print(f"[Checkpoint] 已复制到 Google Drive: {drive_output_filepath}")
        except Exception as e:
            print(f"[Checkpoint] 复制文件到 Google Drive 时出错: {e}")

        # 3. 复制到 GitHub 仓库目录并推送
        try:
            shutil.copyfile(output_filepath, github_qa_filepath)
            print(f"[Checkpoint] 已复制到 GitHub 仓库本地路径: {github_qa_filepath}")

            # 切换到 GitHub 仓库目录进行 git 操作
            os.chdir(github_repo_path)

            !git add data/processed/qa_instructions.jsonl
            !git commit -m "feat: Checkpoint - Add qa_instructions.jsonl (batch {len(qa_instructions)})"
            # 使用 PAT 推送
            !git push https://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@github.com//Datasets-for-Top-Journals-in-Operations-Research-and-Optimization.git main

            print(f"[Checkpoint] 已成功推送到 GitHub (批次: {len(qa_instructions)})\n")

            # 切换回原始工作目录
            os.chdir(original_cwd)

        except Exception as e:
            print(f"[Checkpoint] 推送到 GitHub 时出错: {e}")

# 最终保存所有生成的问答指令到本地文件
with jsonlines.open(output_filepath, mode='w') as writer:
    writer.write_all(qa_instructions)

print(f"已生成 {len(qa_instructions)} 条问答指令并保存到 {output_filepath}")

# 最终复制文件到 Google Drive
try:
    shutil.copyfile(output_filepath, drive_output_filepath)
    print(f"已将 qa_instructions.jsonl 最终版本复制到 Google Drive: {drive_output_filepath}")
except Exception as e:
    print(f"最终复制文件到 Google Drive 时出错: {e}")

# 最终推送到 GitHub，确保所有数据都已同步
try:
    # 复制文件到 GitHub 仓库目录
    shutil.copyfile(output_filepath, github_qa_filepath)
    print(f"最终复制到 GitHub 仓库本地路径: {github_qa_filepath}")

    # 切换到 GitHub 仓库目录进行 git 操作
    os.chdir(github_repo_path)

    !git add data/processed/qa_instructions.jsonl
    !git commit -m "feat: Final push - Update qa_instructions.jsonl"
    !git push https://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@github.com//Datasets-for-Top-Journals-in-Operations-Research-and-Optimization.git main

    print(f"qa_instructions.jsonl 最终版本已推送到 GitHub 仓库。")

    # 切换回原始工作目录
    os.chdir(original_cwd)

except Exception as e:
    print(f"最终推送到 GitHub 时出错: {e}")
````

### 2. 模型选择与微调 (LoRA)

我们将选择一个小型、易于微调的模型，例如 `Qwen/Qwen1.5-0.5B-Chat`。为了高效地进行微调，我们将采用 LoRA (Low-Rank Adaptation) 技术，它只训练模型的一小部分参数，大大减少了计算资源需求。

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer
from datasets import Dataset

# 更新 torchao 库到兼容版本，以解决 ImportError
!pip install torchao>=0.16.0 -q
print("torchao has been updated. Please restart the Colab runtime (Runtime -> Restart runtime) for the changes to take effect, then re-run all cells.")

# 模型路径
model_name = "Qwen/Qwen1.5-0.5B-Chat"

# 1. 配置量化参数 (4-bit quantization) 以节省显存
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=False,
)

# 2. 加载预训练模型
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    trust_remote_code=True,
    device_map="auto", # 自动分配到可用设备 (如 GPU)
)

# 3. 加载分词器
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token # 设置 padding token
tokenizer.padding_side = "right" # 对于因果语言模型，通常将 padding 放在右侧

# 4. 准备模型进行 kbit 训练 (LoRA 训练前的必要步骤)
model = prepare_model_for_kbit_training(model)

# 5. 配置 LoRA 参数
peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.1,
    r=64,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
        "lm_head",
    ], # Qwen1.5 模型的 target modules
)

# 6. 获取 PEFT 模型
model = get_peft_model(model, peft_config)
model.print_trainable_parameters() # 打印可训练参数数量

print("模型和分词器已加载，LoRA 配置已应用。")
````

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from transformers import BitsAndBytesConfig

MAX_SEQ_LEN = 1024
BASE_MODEL  = 'Qwen/Qwen2.5-1.5B-Instruct'

bnb_config = BitsAndBytesConfig(
    load_in_4bit              = True,
    bnb_4bit_quant_type       = 'nf4',
    bnb_4bit_compute_dtype    = torch.float16,   # 强制 fp16
    bnb_4bit_use_double_quant = True,
)

model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config = bnb_config,
    device_map          = 'auto',
    torch_dtype         = torch.float16,
)
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'right'

model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    r              = 16,
    lora_alpha     = 32,
    target_modules = ['q_proj','k_proj','v_proj','o_proj',
                      'gate_proj','up_proj','down_proj'],
    lora_dropout   = 0.05,
    bias           = 'none',
    task_type      = 'CAUSAL_LM',
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
````

````python
from datasets import Dataset

SYSTEM_MSG = (
    '你是一名专业的运筹学与管理科学领域的AI助手，'
    '能够准确解答该领域的理论模型、算法设计、优化方法及管理决策等问题。'
)

def format_sample(record):
    messages = [
        {'role': 'system',    'content': SYSTEM_MSG},
        {'role': 'user',      'content': record['instruction']},
        {'role': 'assistant', 'content': record['output']},
    ]
    text = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=False
    )
    return {'text': text}

train_dataset = Dataset.from_list(train_data).map(format_sample)
print(f'训练集: {len(train_dataset)} 条')
print(train_dataset[0]['text'][:300])
````

````python
import torch, math
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel

SYSTEM_MSG = (
    '你是一名专业的运筹学与管理科学领域的AI助手，'
    '能够准确解答该领域的理论模型、算法设计、优化方法及管理决策等问题。'
)
EVAL_SAMPLE = eval_data[:100]

def compute_perplexity(mdl, tok, records, max_len=512):
    mdl.eval()
    total_loss, count = 0.0, 0
    with torch.no_grad():
        for r in records:
            full_msgs   = [
                {'role':'system',    'content': SYSTEM_MSG},
                {'role':'user',      'content': r['instruction']},
                {'role':'assistant', 'content': r['output']},
            ]
            prompt_msgs = [
                {'role':'system', 'content': SYSTEM_MSG},
                {'role':'user',   'content': r['instruction']},
            ]
            full_text   = tok.apply_chat_template(full_msgs,   tokenize=False, add_generation_prompt=False)
            prompt_text = tok.apply_chat_template(prompt_msgs, tokenize=False, add_generation_prompt=True)

            input_ids  = tok(full_text,   return_tensors='pt', truncation=True, max_length=max_len).input_ids.cuda()
            prompt_len = tok(prompt_text, return_tensors='pt').input_ids.shape[1]

            labels = input_ids.clone()
            labels[:, :prompt_len] = -100

            loss = mdl(input_ids, labels=labels).loss
            if not torch.isnan(loss):
                total_loss += loss.item()
                count += 1
    return math.exp(total_loss / max(count, 1))

# 微调后困惑度
print('计算微调后困惑度...')
ppl_finetuned = compute_perplexity(model, tokenizer, EVAL_SAMPLE)
print(f'微调后 PPL: {ppl_finetuned:.2f}')
````

````python
bnb_config = BitsAndBytesConfig(
    load_in_4bit           = True,
    bnb_4bit_quant_type    = 'nf4',
    bnb_4bit_compute_dtype = torch.float16,
)

print('加载基座模型...')
base_model = AutoModelForCausalLM.from_pretrained(
    'Qwen/Qwen2.5-1.5B-Instruct',
    quantization_config = bnb_config,
    device_map          = 'auto',
    torch_dtype         = torch.float16,
)
base_tok = AutoTokenizer.from_pretrained('Qwen/Qwen2.5-1.5B-Instruct')
base_tok.pad_token = base_tok.eos_token

ppl_base = compute_perplexity(base_model, base_tok, EVAL_SAMPLE)
print(f'基座模型 PPL:  {ppl_base:.2f}')
print(f'微调后 PPL:    {ppl_finetuned:.2f}')
print(f'困惑度降低:    {ppl_base - ppl_finetuned:.2f}  ({(ppl_base-ppl_finetuned)/ppl_base*100:.1f}%)')

del base_model, base_tok
torch.cuda.empty_cache()
````

````python
import os, json, time
from openai import OpenAI

os.environ['DEEPSEEK_API_KEY'] = 'XXXXXXXXXXXXXXXXXXXXXXXXXXX'
judge_client = OpenAI(api_key=os.environ['DEEPSEEK_API_KEY'], base_url='https://api.deepseek.com')

# 加载基座模型用于生成对比答案
bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_quant_type='nf4', bnb_4bit_compute_dtype=torch.float16)
base_model = AutoModelForCausalLM.from_pretrained('Qwen/Qwen2.5-1.5B-Instruct', quantization_config=bnb_config, device_map='auto', torch_dtype=torch.float16)
base_tok = AutoTokenizer.from_pretrained('Qwen/Qwen2.5-1.5B-Instruct')
base_tok.pad_token = base_tok.eos_token

def generate_answer(mdl, tok, question, max_new_tokens=300):
    msgs   = [{'role':'system','content':SYSTEM_MSG}, {'role':'user','content':question}]
    prompt = tok.apply_chat_template(msgs, tokenize=False, add_generation_prompt=True)
    inputs = tok(prompt, return_tensors='pt').to('cuda')
    with torch.no_grad():
        out = mdl.generate(**inputs, max_new_tokens=max_new_tokens,
                           temperature=0.3, do_sample=True, pad_token_id=tok.eos_token_id)
    return tok.decode(out[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True).strip()

JUDGE_SYSTEM = """你是运筹学与管理科学领域的专家评委，对两个AI回答评分。
评分维度(各25分，总100)：专业准确性、推理深度、完整性、清晰度。
A=基座模型，B=微调模型。
只输出JSON，不要其他内容：{"score_A":<0-100>,"score_B":<0-100>,"winner":"A" or "B" or "tie","reason":"<20字>"}"""

def llm_judge(question, ans_a, ans_b):
    user = f'问题：{question}\n\n回答A：\n{ans_a}\n\n回答B：\n{ans_b}'
    for _ in range(2):
        try:
            resp = judge_client.chat.completions.create(
                model='deepseek-chat',
                messages=[{'role':'system','content':JUDGE_SYSTEM},{'role':'user','content':user}],
                temperature=0.1, timeout=30,
            )
            raw = resp.choices[0].message.content.strip().strip('`').lstrip('json').strip()
            return json.loads(raw)
        except: time.sleep(2)
    return None

JUDGE_SAMPLE = eval_data[:30]
results = []
for i, r in enumerate(JUDGE_SAMPLE):
    q        = r['instruction']
    ans_base = generate_answer(base_model, base_tok, q)
    ans_ft   = generate_answer(model, tokenizer, q)
    judgment = llm_judge(q, ans_base, ans_ft)
    if judgment:
        results.append({'question':q, 'ans_base':ans_base, 'ans_finetuned':ans_ft,
                        'score_base':judgment['score_A'], 'score_finetuned':judgment['score_B'],
                        'winner':judgment['winner'], 'reason':judgment.get('reason','')})
    if (i+1) % 5 == 0:
        print(f'已评估 {i+1}/{len(JUDGE_SAMPLE)}')
    time.sleep(0.5)

print(f'完成，有效评判 {len(results)} 条')
````
